```python
import matplotlib.pyplot as plt
from arenz_group_python.ec_data import EC_Data
from pathlib import Path
import sys
```

```python
#Find CV files to plot
files = list(Path().glob('Steps*.tdms'))

#Load all data files
data = []
for path_to_file in files:
    data.append(EC_Data(path_to_file))
```

```python
#Plot CV files
fig = plt.figure()
plt.suptitle("Steps")
b = fig.subplots(3)
for x in data:
    x.plot("Time","E", p=b[0])
    x.plot("Time","i", p=b[1])
    x.plot("E","i", p=b[2])
```
