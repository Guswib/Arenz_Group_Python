```python
import matplotlib.pyplot as plt
#import arenz_group_python
from arenz_group_python.ec_data_module import EC_Data
from pathlib import Path
import sys
```

```python
#Find CV files to plot
files = list(Path().glob('Steps*.tdms'))

#Load all data files
data = [EC_Data() for _ in range(len(files))]
#data = []
i=0
for path_to_file in files:
    #data.append(EC_Data(path_to_file))
    data[i] = EC_Data(path_to_file)
    i=i+1
```

```python
#Plot CV files
fig = plt.figure()
plt.suptitle("Steps")
b = fig.subplots(3)
b[0]
for x in data:
    x.plot("Time","E", plot=b[0])
    x.plot("Time","i", plot=b[1])
    x.plot("E","i", plot=b[2])
```
