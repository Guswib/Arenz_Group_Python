```python
from PIL import Image
from PIL.TiffTags import TAGS,TAGS_V2
 
def load_image(file_path):
    try:
        # Open the image file
        img = Image.open(file_path)
        # Extract metadata
        metadata = img.tag_v2
        for i, page,in enumerate(metadata):
            #print(i,page)
            tag_name = str(TAGS_V2.get(i,page))
            print(i,page,tag_name)
            
        # Print all metadata to inspect
        #print(metadata)
        a=[]
        b=[]
        for tag, value in metadata.items():
            tag_name = str(TAGS.get(tag, tag))
            if tag_name.startswith("P*i"):
                print("FOUND", tag_name, value)
            #print(f"Tag: {tag}, Value: {value}")
            a.append(tag_name)
            b.append(value)
        #for i in a:
        #    print(i)
        print(len(b))
        print(a[6])
        s=str(b[6]).split("\r",500)
        k=0
        for a in s:
            qq = a
        
            if qq.strip().startswith("AP_PIXEL_SIZ"):
                print(k,"\tqqq",a)
            k=k+1
        print("values############################")
        k=0
        for i in b:
            #if str(i).startswith("P"):
                print("AAAAAAAAAAAAAA",k, "\n",str(i))  
                k=k+1          
    except Exception as e:
        print(f"Error loading image: {e}")

# Specify the file path
file_path = 'C:\\Users\\gustav.wiberg\\Documents\\GitHub\\Python\\Arenz_Group_Python\\tests\\data\\set1\\LED-DG008-0-AuNPs_02.tif'  # Replace with your file path
load_image(file_path)
```

```python
import tifffile
 
def extract_image_pixel_size(file_path):
    try:
        with tifffile.TiffFile(file_path) as tif:
            for i, page in enumerate(tif.pages):
                if 34118 in page.tags:
                    description = page.tags[34118].value
 
                    if isinstance(description, dict):
                        # Check for 'ap_image_pixel_size'
                        if 'ap_image_pixel_size' in description:
                            value = description['ap_image_pixel_size']
                            if isinstance(value, tuple) and len(value) == 3:
                                label, measurement, unit = value
                                print(f"ap_image_pixel_size: {label} = {measurement} {unit}")
                            else:
                                print(f"ap_image_pixel_size: {value}")
                        else:
                            print("'ap_image_pixel_size' not found in ImageDescription.")
                    else:
                        print(f"ImageDescription is of type: {type(description)}")
                else:
                    print("ImageDescription tag (34118) not found.")
    except Exception as e:
        print(f"An error occurred: {e}")
 

    # Replace the file path below with the path to your TIFF file
file_path = 'C:\\Users\\gustav.wiberg\\Documents\\GitHub\\Python\\Arenz_Group_Python\\tests\\data\\set1\\LED-DG008-0-AuNPs_02.tif'  # Replace with your file path
extract_image_pixel_size(file_path)
```
